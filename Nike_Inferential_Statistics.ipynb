###### Q 1. Is there a difference in usage level between male and female customers?
##### Answer: We will use Chi-square test to find this answer. The Chi-Square Test of Independence determines whether there is an association between categorical variables (i.e., whether the variables are independent or related).

##### Hnull: There is no association between two categorical variables
##### H Alt: Both the categorial variables are associated with other

###### It also gives us cramer'V coeffcient , which is measure of level of association between two variables. It is between 0 and 1. Values closer to 1 indicates higher association between the variables.
    
##### In the following example, p-value is 0.04, which is less than 0.05. It indicates that there is not sufficient evidence for not rejecting the null hypothesis. So, we will reject the null hypothesis. We can conclude that there is different level of usage among male and female customers. 

```python
crosstab, test_results, expected = rp.crosstab(data["sex"], data["usage"],
                                               test="chi-square",
                                               expected_freqs= True,
                                               prop= "cell")

print(test_results)
```

```python
p=test_results['results'][1]
alpha = 0.05

print("H0: both the nominal varibales are not associated with each other")
if p > alpha:
    print('Result: These two nominal variables are not associated (fail to reject H0)')
else:
    print('Result: These two nominal variables are associated (reject H0)')
```

###### Q 4.Assume that awareness toward Nike was measured on an ordinal scale rather than an interval scale. Do males and females differ in their awareness towards Nike?

###### Answer: As awareness is measured on ordinal scale now. so we can not apply independent sample t-test. It assumes the variable to be numeric. We will conduct Mann Whitney U test to compare if male and females differs in their awareness for nike.

###### Hnull:Both male and female have equal ranking for awareness 
###### H Alt: Male and female differ in their ranking with respect to awareness. 

###### In the following example, we find a p-value of 0.009. It is less than 0.05. Hence, we will reject the null hypothsis and conclude that male and female differ in their level of awareness.

###### To further check if , male have higher awareness level than female customers, we will run the mann whitney test againg with one more parameter alaternative=\"greater\". P-value less than 0.05 indicates that male has higher awareness than female cusotmers. 

```python
## Non-Parametric Independent sample test-- Mann-Whitney U test
stats.mannwhitneyu(Awareness_male,Awareness_Female)
```

```python
stats.mannwhitneyu(Awareness_Female,Awareness_male,alternative='less')

# H null: Male and female have equal wareness.
# H alt: Male have higher awareness than female customers.

## you can also pass alternative= \"less\" for checking if male customers have lower awareness than female customers.
```

In a paired sample t-test, each subject or entity is measured twice, resulting in pairs of observations.

##### Q. 6 Assume that awareness and loyalty were measure on ordinal scale rather than an interval scale. Do the respondents have higher awareness for Nike than loyalty for Nike?

##### Answer: We will use wilcoxon singed rank t-test. The Wilcoxon signed-rank test is a non-parametric statistical hypothesis test used to compare two related samples, matched samples, or repeated measurements on a single sample to assess whether their population mean ranks differ (i.e. it is a paired difference test). 

###### Hnull: Rank of awareness and loyalty are same among customers.
###### H Alt: Rank of awareness is higher than that of loyalty for nike.

##### to Check if level of awareness is greater than level of preferecne, we can put alternative=\"greater\" argument in function. To check if awareness is lesser than preference, we can put  alternative=\"less\". To check if both are different, we can put alternative=\"two-sided\". Default option is \"two-sided\". 

##### we can find that p-value is 0.66, which is greater that 0.05. Hence, we will not reject the null hypothesis. We will conclude that customers do not differ in their awareness and preference for nike. 

```python
## Non-Parametric wilxcon paired t-test--Wilcoxon Signed Rank Test
stats.wilcoxon(data['awareness'], data['preference'],alternative='greater')
```

###### Q. 7 Do the various types of customers in terms of usage differ in their awareness for Nike? Their attitude for Nike? Their loyalty for Nike?
###### Answer: Usage level has three groups heavy, medium, and light. As it has more than two groups, we can not apply independent sample t-test to check the group differences. We will conduct one-way analysis of variance (ANOVA) in Python

###### Hnull: There is no difference in mean value of awareness across heavy, medium, and light users.
###### H alt:  Mean of awareness is different across the heavy, medium, and light users.

###### First, we can run group by command to find the average score of awareness for heavy, light, and medium users. We can see that awareness mean is different for these three groups. We will run anova analysis in python. Getting p-value less than 0.05 indicates that mean of awareness of different across heavy, light, and medium usage customers.

###### To compare the mean difference among different pairs of usage level, we can use tukey's post hoc analysis. Let us interpret the results. First line represnts the difference between light and heavy group. Mean diff is -3.41 with p-value = 0.001. As p value is less than 0.05, we conclude that mean awareness for heavy and light users are significant different and mean awareness of heavy users is 3.41 higher than that of light users. 

###### To check the difference in attitude and loyalty among various usage level customers, please run the same analysis for attitude and loyalty variable as well. 

```python
data.groupby(['usage'])['awareness'].mean()
```

```python
# stats f_oneway functions takes the groups as input and returns F and P-value
Attribute = 'awareness'
X1=data.loc[data["usage"]=='Heavy',Attribute]
X2=data.loc[data["usage"]=='Medium',Attribute]
X3=data.loc[data["usage"]=='Light',Attribute]
fvalue, pvalue = stats.f_oneway(X1,X2,X3)
print(fvalue, pvalue)
```

```python
## for conducting tukey's post hoc analysis.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
m_comp = pairwise_tukeyhsd(endog=data['attitude'], groups=data['usage'], alpha=0.05)
print(m_comp)
```
